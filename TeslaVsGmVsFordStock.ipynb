```python
import requests
import yfinance as yf
import pandas as pd
```

```python
Gm=yf.Ticker("GM")
Ford=yf.Ticker("F")
Tesla=yf.Ticker('TSLA')
```

```python
gm_info=Gm.history(period='10y')
ford_info=Ford.history(period='10y')
tesla_info=Tesla.history(period='10y')
```

```python
sl=[gm_info,ford_info,tesla_info]
for i in sl:
  i.reset_index(inplace=True)
```

```python
sl=[gm_info,ford_info,tesla_info]
cnt=0
for i in sl:
  if cnt==0:
    print("----------------------------------GM MOTOR---------------------------------")
  if cnt==1:
    print("----------------------------------FORD MOTOR---------------------------------")
  if cnt==2:
    print("----------------------------------TESLA MOTOR---------------------------------")
  cnt+=1
  print(i.head())
  print(i.describe())
  print(i.tail(),"\n\n")
```

```python
import matplotlib.pyplot as plt
sl=[gm_info,ford_info,tesla_info]
count2=0
for i in sl:
  if count2==0:
    ttl="GM MOTORS"
  if count2==1:
    ttl="FORD MOTORS"
  if count2==2:
    ttl="TESLA MOTORS"
  count2=count2+1
  i.plot(x="Date",y="High",title=ttl)
```
